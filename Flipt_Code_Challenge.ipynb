## Generalized plotting function to plot distributions

```python
def plotDistribution(plottingVariable):
    fig = plt.figure(figsize=(12,9))
    plt.title('Distribution') 
    sns.distplot(plottingVariable, bins=50)
```

## Function to test distribution

```python
def normalityTest(testData):
    
    # D'Agostino and Pearson's Test
    data = testData
    # normality test
    stat, p = normaltest(data)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print('Sample looks Gaussian (fail to reject H0)')
    else:
        print('Sample does not look Gaussian (reject H0)')
```

## 5. a) For the same subset above, randomly assign one of the three saving percentages (10%/15%/20%) to each row, and calculate the total cost after savings in a new column: 
### total cost after savings = total cost * (1 - savings %)

```python
mid_trend_saving = mid_trend

#Randomly assigning 10%, 15%, and 20% to the rows with 33.33% probability
mid_trend_saving['savings'] = mid_trend_saving['simulation'].transform(lambda x: np.random.choice([0.1,0.15,0.2]))
```

```python
mid_trend_saving.head()
```

```python
#Applying the formula on the dataframe to calculate cost after savings

mid_trend_saving['total_cost_after_savings'] = mid_trend_saving.total_cost * (1-mid_trend_saving.savings)
```
